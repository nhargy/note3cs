```python
# custom analysis tools
from analysis.scan import scan
from analysis.tools import extract_species
from analysis.tools import muon_filter
from analysis.tools import muon_filter_by_parts
from analysis.tools import norm_df
from analysis.tools import diff_df
from analysis.tools import ambience_435

# useful python libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd
from prettytable import PrettyTable

plt.rcParams['figure.figsize'] = 24, 10
plt.style.use('ggplot')

STUDY = 'master_neutrons'

# Get the current directory where the Jupyter Notebook is located

current_directory = get_ipython().run_line_magic('pwd', '')

# Extract the local path to 3CS folder
local_path        = current_directory.split('repo_note3cs')[0]

# Define path to study within data3cs folder
path_to_study     = f'repo_data3cs/data3cs/studies/{STUDY}'

# Final path
path              = f'{local_path}{path_to_study}'

# Irradiation group
LiF_irrad                = extract_species(path, 'LiF', ext='irrad') 
LiF_irrad_samples = ['LiF_B1_131', 'LiF_B2_145', 'LiF_B2_157', 'LiF_B2_158']

# Control group
LiF_ctrl                     = extract_species(path, 'LiF', ext='ctrl') 
LiF_ctrl_samples      = ['LiF_B1_130', 'LiF_B2_143', 'LiF_B2_155']
```

```python
# Gaussian base functions

def Gaussian(x, A, m, s, offset = 0):
    return A * np.exp( -0.5 * ((x - m)**2) / (s**2) ) + offset

def Gaussian2(x, A1, m1, s1, A2, m2, s2):
    return Gaussian(x,A1,m1,s1) + Gaussian(x,A2,m2,s2)

def Gaussian3(x,A1, m1, s1, A2, m2, s2, A3, m3, s3):
    return Gaussian(x,A1,m1,s1) + Gaussian(x,A2,m2,s2) + Gaussian(x,A3,m3,s3)

def remove_nans(x,y, err):
    size = len(y)
    indexes = []
    for i in range(size):
        if np.isnan(y[i]):
            indexes.append(i)

    x_clean = np.delete(x, indexes)
    y_clean = np.delete(y, indexes)
    err_clean = np.delete(err, indexes)
    
    return x_clean, y_clean ,err_clean

ratio_260 = 12.743
ratio_435 = 11.703

BGU_gamma_dose =  35 #mGy
WIS_gamma_dose = 330 #mGy
```

```python
wl           = '435'

tot_arr =  []
gam_arr = []
for sample in LiF_irrad_samples:

    s1_lif = norm_df(LiF_irrad, f'{sample}_B', 3, wl, 'ambience_435', ratio_435)
    s2_lif = norm_df(LiF_irrad, f'{sample}_B', 4, wl, 'ambience_435', ratio_435)
    ds_lif = diff_df(s2_lif,s1_lif)
    
    x = ds_lif['em_wl']; y = ds_lif['count']; err = ds_lif['error']
    x, y, e = remove_nans(ds_lif['em_wl'], ds_lif['count'], ds_lif['error'])
    
    combined_error = np.round(np.sqrt(np.sum(err**2)), 2)
    area                        = np.round(np.trapz(y),2)

    gam_err                = np.round(combined_error / BGU_gamma_dose , 2)
    gam_area               = np.round(area / BGU_gamma_dose, 2)
    
    #print(f'{sample} Total photon count = {area}  +-({combined_error})')

    tot_arr.append((area, combined_error))
    gam_arr.append((gam_area, gam_err))

print(' * * * * * * * * * * * * * * * * * * * * * * * * Neutron (BGU) Irradiation Campaign * * * * * * * * * * * * * * * * * * * * * * * * ')
print()
print('Irradiation Group AmBe (Total gamma dose ~ 35mGy):')

# Create a PrettyTable object
table = PrettyTable()

# Define the columns
table.field_names = [r"Sample", "Total Photons/sec/microWatt (err)", "Photons/sec/microWatt per mGy gamma (err)"]

# Add data to the table
table.add_row([LiF_irrad_samples[0], f'{tot_arr[0][0]} ({tot_arr[0][1]})', f'{gam_arr[0][0]} ({gam_arr[0][1]})'])
table.add_row([LiF_irrad_samples[1], f'{tot_arr[1][0]} ({tot_arr[1][1]})', f'{gam_arr[1][0]} ({gam_arr[1][1]})'])
table.add_row([LiF_irrad_samples[2], f'{tot_arr[2][0]} ({tot_arr[2][1]})', f'{gam_arr[2][0]} ({gam_arr[2][1]})'])
table.add_row([LiF_irrad_samples[3], f'{tot_arr[3][0]} ({tot_arr[3][1]})', f'{gam_arr[3][0]} ({gam_arr[3][1]})'])

# Print the table
print(table)

avr0 = np.round(np.sum(gam_arr) / 4 , 1)
print(f'Average ~ {avr0} Photons/sec/microWatt per mGy gamma')


# --- ctrl --- #

tot_arr =  []
gam_arr = []
for sample in LiF_ctrl_samples:

    s1_lif = norm_df(LiF_ctrl, f'{sample}_B', 3, wl, 'ambience_435', ratio_435)
    s2_lif = norm_df(LiF_ctrl, f'{sample}_B', 4, wl, 'ambience_435', ratio_435)
    ds_lif = diff_df(s2_lif,s1_lif)
    
    x = ds_lif['em_wl']; y = ds_lif['count']; err = ds_lif['error']
    x, y, e = remove_nans(ds_lif['em_wl'], ds_lif['count'], ds_lif['error'])
    
    combined_error = np.round(np.sqrt(np.sum(err**2)), 3)
    area                        = np.round(np.trapz(y),3)

    gam_err                = np.round(combined_error / BGU_gamma_dose , 4)
    gam_area               = np.round(area / BGU_gamma_dose, 4)
    
    #print(f'{sample} Total photon count = {area}  +-({combined_error})')

    tot_arr.append((area, combined_error))
    gam_arr.append((gam_area, gam_err))

print()
print()
print('Control Sample:')

# Create a PrettyTable object
table = PrettyTable()

# Define the columns
table.field_names = [r"Sample", "Total Photons/sec/microWatt (err)", "Photons/sec/microWatt per mGy gamma (err)"]

# Add data to the table
table.add_row([LiF_irrad_samples[0], f'{tot_arr[0][0]} ({tot_arr[0][1]})', f'{gam_arr[0][0]} ({gam_arr[0][1]})'])
table.add_row([LiF_irrad_samples[1], f'{tot_arr[1][0]} ({tot_arr[1][1]})', f'{gam_arr[1][0]} ({gam_arr[1][1]})'])
table.add_row([LiF_irrad_samples[2], f'{tot_arr[2][0]} ({tot_arr[2][1]})', f'{gam_arr[2][0]} ({gam_arr[2][1]})'])

# Print the table
print(table)

avr1 = np.round(np.sum(gam_arr) / 3 , 2)
print(f'Average ~ {avr1} Photons/sec/microWatt per mGy gamma')

print()
print()
print()
print()

print(' * * * * * * * * * * * * * * * * * * * * * * * * Gamma-only (WIS) Irradiation Campaign * * * * * * * * * * * * * * * * * * * * * * * * ')
print()

print('Co60 Irradiation (Total gamma dose ~ 330mGy):')

# lif_agreement_2
xg, yg_145, yg_157 = np.load('agreement_lif_2.npy')

area_145 = np.round(np.trapz(yg_145) , 4)
area_157 = np.round(np.trapz(yg_157), 4)

area_145_gam = np.round(area_145 / WIS_gamma_dose, 4)
area_157_gam = np.round(area_157 / WIS_gamma_dose, 4)

# Create a PrettyTable object
table = PrettyTable()

# Define the columns
table.field_names = [r"Sample", "Total Photons/sec/microWatt", "Photons/sec/microWatt per mGy gamma"]

# Add data to the table
table.add_row(['LiF_B2_145', area_145, area_145_gam])
table.add_row(['LiF_B2_157' , area_157, area_157_gam])

# Print the table
print(table)

avr2 = np.round((area_145_gam + area_157_gam) / 2, 2)
print(f'Average ~ {avr2} Photons/sec/microWatt per mGy gamma')


print()
print()
print()

print(' * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * ')
print()
excess = np.round(avr0/avr2); perc = np.round(1/excess, 4) * 100
print(f'Neutron excess factor ~ {excess};  {perc}% of the signal can be attributed to gammas.')
```
