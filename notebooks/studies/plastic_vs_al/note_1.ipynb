```python
import h5py as h5
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = 18,8
```

```python
def extract_data(path, ori = "0", wl='450'):
    
    f = h5.File(path)
    gp = f['spectra'][wl][ori]['15.0sec']
    
    sp_data = gp['sp_data']; bg_data = gp['bg_data']; powers = gp['power_samples']
    
    return sp_data, bg_data, powers
```

```python
path_to_empty_plastic  = '/home/hargy/Documents/3CS/repo_data3cs/data3cs/studies/plastic_vs_al/coll1/EMPTY_PLASTIC.hdf5' #EP
path_to_sp_plastic    = '/home/hargy/Documents/3CS/repo_data3cs/data3cs/studies/plastic_vs_al/coll1/SP_PLASTIC.hdf5'     #SP
path_to_empty_al      = '/home/hargy/Documents/3CS/repo_data3cs/data3cs/studies/plastic_vs_al/coll2/EMPTY_AL.hdf5'       #EA
path_to_sp_al         = '/home/hargy/Documents/3CS/repo_data3cs/data3cs/studies/plastic_vs_al/coll2/SP_AL.hdf5'          #SA
```

```python
sp_ep, bg_ep, pw_ep = extract_data(path_to_empty_plastic, ori = 'unk', wl = '300')
sp_sp, bg_sp, pw_sp = extract_data(path_to_sp_plastic, ori = '0', wl = '300')
sp_ea, bg_ea, pw_ea = extract_data(path_to_empty_al, ori = 'unk', wl = '300')
sp_sa, bg_sa, pw_sa = extract_data(path_to_sp_al, ori = '0', wl = '300')
```

```python
it = 1; wl = sp_ep[it]['em_wl'] 
plt.plot(wl, sp_ep[it]['count'], label = 'EP')
plt.plot(wl, sp_sp[it]['count'], label = 'SP')
plt.plot(wl, sp_ea[it]['count'], label = 'EA')
plt.plot(wl, sp_sa[it]['count'], label = 'SA')
plt.legend()
```
