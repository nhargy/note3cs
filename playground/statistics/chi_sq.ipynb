```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = 24, 10
plt.style.use('dark_background')
```

```python
def chi_squared(array):
    sq_array = np.square(array)
    chi_sq      = np.sum(sq_array)
    return chi_sq
```

```python
mean = 0; std_dev = 1; size = 3
its = 10000

# Generate all random values at once
random_values = np.random.normal(mean, std_dev, (its, size))

# Calculate chi-squared values for each set of random values
chi_squared_values = np.sum((random_values - mean) ** 2 / std_dev ** 2, axis=1)

# Convert the result to a list if needed
arr = chi_squared_values.tolist()
```

```python
plt.hist(arr, bins = 100, density = True);
```
